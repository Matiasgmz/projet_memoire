```python
import pandas as pd
```

```python
df_results = pd.read_csv('data/results.csv')
df_results
```

```python
df_results.isna().sum()
```

```python
df_races = pd.read_csv('data/races.csv')
df_races = df_races.drop(columns=['time'])
df_races = df_races.rename(columns={'name': 'circuit_name'})
df_races
```

```python
df_races.isna().sum()
```

```python
df_drivers = pd.read_csv('data/drivers.csv')
df_drivers = df_drivers.drop(columns=['url'])
df_drivers = df_drivers.rename(columns={'nationality': 'driver_nationality'})
df_drivers = df_drivers.rename(columns={'number': 'driver_number'})
df_drivers
```

```python
df_drivers.isna().sum()
```

```python
df_constructors = pd.read_csv('data/constructors.csv')
df_constructors = df_constructors.drop(columns=['url'])
df_constructors = df_constructors.rename(columns={'name': 'constructor_name'})
df_constructors = df_constructors.rename(columns={'nationality': 'constructor_nationality'})
df_constructors
```

```python
df_drivers.isna().sum()
```

```python
df_circuits = pd.read_csv('data/circuits.csv')
df_circuits = df_circuits.drop(columns=['url'])
df_circuits
```

```python
df_circuits.isna().sum()
```

```python
merged_df = pd.merge(df_results, df_races, on='raceId')
merged_df = pd.merge(merged_df, df_drivers, on='driverId')
merged_df = pd.merge(merged_df, df_constructors, on='constructorId')
merged_df = pd.merge(merged_df, df_circuits, on='circuitId')
merged_df.head()
```

```python
merged_df.dtypes
```

```python
merged_df.columns
```

```python
columns_to_use = ['driverId', 'constructorId', 'grid',  'rank', 'year', 'round', 'circuitId']
df = merged_df[columns_to_use]
df
```

```python
df_filtered = df.query('2021 <= year <= 2023')
df_filtered
```
